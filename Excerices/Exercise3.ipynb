```python
#Exercise 3

from colorama import Fore, Style

#function to detect right selected genre and not duplicate selected
def is_in_genre(selected_genre,*othergenre) :
    for i in range(len(othergenre)) :
        if selected_genre==othergenre[i] :
            print(Fore.BLUE +"*** Error : You have chosen a duplicate genre ***") 
            print(Style.RESET_ALL)
            return False         
    if selected_genre in genre_list :
            return True
    else :
            print(Fore.RED +"*** Error : Please Select genre From The list Above ***") 
            print(Style.RESET_ALL)
            return False

genre_list=['Horror','Romance','Comedy','History','Adventure','Action']
genre_Grade=[0,0,0,0,0,0]

while True :
    try :
        x=int(input('How many people are involved in this question?'))
        break
    except :
        print(Fore.RED + '************* ERROR : Please Enter a Number *************')
        print(Style.RESET_ALL)      
    

print('\nList of genre is : '," ".join(genre_list))   

for i in range(x) :
    print(Fore.GREEN," ")
    name=input('\nPlease Enter Your Name :')
    print(Style.RESET_ALL)
    while True :
        genre1=input('Dear '+name+' Please Enter your First Favourite genre: ').lower().title()
        if is_in_genre(genre1,"") :
            genre_Grade[genre_list.index(genre1)]+=1
            break
    while True :
        genre2=input('Dear '+name+' Please Enter your Second Favourite genre: ').lower().title()
        if is_in_genre(genre2,genre1):
            genre_Grade[genre_list.index(genre2)]+=1
            break
    while True :
        genre3=input('Dear '+name+' Please Enter your Third Favourite genre: ').lower().title()
        if is_in_genre(genre3,genre1,genre2) :
            genre_Grade[genre_list.index(genre3)]+=1
            break

#create a tuple
Final_result=list(zip(genre_list,genre_Grade))

#sort the tuple
Final_result.sort(key=lambda x: x[0] ,reverse=False)
Final_result.sort(key=lambda x: x[1] ,reverse=True)

#print Final Result
print('\n')
for item in Final_result :
    print(item[0] , ':' , item[1])
```
